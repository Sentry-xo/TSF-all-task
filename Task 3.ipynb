## Exploratory Data Analysis

```python
#corelation matrix
plt.figure(figsize=(7,6))
correlation=d.corr()
sns.heatmap(correlation,annot=True,cmap='OrRd')
```

## Lineplot

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=d)
plt.legend('Profit')
```

```python
plt.figure(figsize=(30,13))
plt.plot(d['Profit'], color = 'blue')
plt.grid(alpha=0.2)
```

```python
d.corr()
```

## Distribution Plot

```python
fig, axs = plt.subplots(ncols=2, nrows=2,figsize=(10,10))
sns.distplot(d['Sales'], color = 'blue', ax=axs[0][0])
sns.distplot(d['Profit'], color = 'orange', ax=axs[0][1])
sns.distplot(d['Quantity'], color = 'green', ax=axs[1][0])
sns.distplot(d['Discount'], color = 'red', ax=axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 18)
axs[0][1].set_title('Profit Distribution', fontsize = 18)
axs[1][0].set_title('Quantity Distribution', fontsize = 18)
axs[1][1].set_title('Discount Distribution', fontsize = 18)
plt.show()
```

## State wise analysis of Profit,Discount,Sale

```python
d['State'].value_counts().head(10)
```

```python
d_state = d.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
d_state.head(10)
```

```python
d_state1=d_state.sort_values('Profit')
```

```python
d_state1[['Profit']].plot(kind='bar',figsize=(20,8), color='black')
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per State')
plt.xlabel('States')
plt.show()
```
